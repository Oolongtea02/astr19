```python
#exponential distribution
import numpy as np
import matplotlib.pyplot as plt

scale = 1.0

random_numbers_exponential = np.random.exponential(scale, 1000)

plt.hist(random_numbers_exponential, bins=100, color='green', alpha=1)
plt.title('Histogram of Random Numbers (Exponential Distribution)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
