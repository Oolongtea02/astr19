```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
f, ax = plt.subplots(1, 2, figsize = (4, 4))

ax[0].plot(x, sin_function(x), label='sin(x)',color='purple')
ax[0].set_title('sin(x) vs. x')
ax[0].set_xlabel('x')
ax[0].set_ylabel('sin(x)')

ax[1].plot(x, cos_function(x), label='cos(x)',color='blue')
ax[1].set_title('cos(x) vs. x')
ax[1].set_xlabel('x')
ax[1].set_ylabel('cos(x)')

# Show the plot
plt.show()
```
